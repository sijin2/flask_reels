```python
# ì‚¬ì´í‚·ëŸ°ì˜ CountVectorizer ë¥¼ í†µí•´ ë²¡í„°í™”
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    analyzer = 'word', # ìºë¦­í„° ë‹¨ìœ„ë¡œ ë²¡í„°í™” í•  ìˆ˜ë„ ìˆìŒ
    tokenizer = None, # í† í¬ë‚˜ì´ì €ë¥¼ ë”°ë¡œ ì§€ì •í•´ ì¤„ ìˆ˜ë„ ìˆìŒ
    preprocessor = None, # ì „ì²˜ë¦¬ ë„êµ¬
    stop_words = None, # ë¶ˆìš©ì–´ nltkë“±ì˜ ë„êµ¬ë¥¼ ì‚¬ìš©í•  ìˆ˜ë„ ìˆìŒ
    min_df = 2, # í† í°ì´ ë‚˜íƒ€ë‚  ìµœì†Œ ë¬¸ì„œ ê°œìˆ˜ë¡œ ì˜¤íƒ€ë‚˜ ìì£¼ ë‚˜ì˜¤ì§€ ì•ŠëŠ” íŠ¹ìˆ˜í•œ ì „ë¬¸ìš©ì–´ ì œê±°ì— ì¢‹ìŒ 
    ngram_range=(0, 6), # BOWì˜ ë‹¨ìœ„ ê°¯ìˆ˜ì˜ ë²”ìœ„ë¥¼ ì§€ì •
    max_features = 2000 # ë§Œë“¤ í”¼ì²˜ì˜ ìˆ˜, ë‹¨ì–´ì˜ ìˆ˜
    )
vectorizer
```

```python
feature_vector = vectorizer.fit_transform(df['hash'])
feature_vector = feature_vector
feature_vector.shape
```

```python
vocab = vectorizer.get_feature_names_out()
vocab= vocab
print(len(vocab))
vocab[:10]
```

```python
# TfidfTransformer ë¥¼ ë¶ˆëŸ¬ì™€ì„œ ê°€ì¤‘ì¹˜ë¥¼ ì£¼ì–´ ë²¡í„°í™”
# transformer ë¼ëŠ” ë³€ìˆ˜ë¡œ ì €ì¥í•˜ê³  ì¬ì‚¬ìš©í•©ë‹ˆë‹¤.
from sklearn.feature_extraction.text import TfidfTransformer
tfidftrans = TfidfTransformer(smooth_idf=False)
tfidftrans
```

```python
feature_tfidf = tfidftrans.fit_transform(feature_vector)
feature_tfidf.shape
```

```python
tfidf_freq = pd.DataFrame(feature_tfidf.toarray(), columns=vocab)
tfidf_freq.head() 
```

```python
df_tfidf = pd.DataFrame(tfidf_freq.sum())
df_tfidf_top = df_tfidf.sort_values(by=0, ascending=False)
df_tfidf_top.head(10)
```

```python
# ì¤‘ê°„ì— ìƒëµë˜ëŠ” ë‹¨ì–´ë¥¼ ìì„¸íˆ ë³´ê³ ì í•  ë•Œ
for t in df_tfidf_top.index[:30]:
    print(t)
```

```python
from sklearn.cluster import KMeans
from tqdm import trange
inertia = []

start = 10
end = 100

# ì ì ˆí•œ í´ëŸ¬ìŠ¤í„°ì˜ ê°¯ìˆ˜ë¥¼ ì•Œê¸° ìœ„í•´ inertia ê°’ì„ êµ¬í•¨
# trange ë¥¼ í†µí•´ ì‹œì‘ê³¼ ë ê°’ì„ ì§€ì •í•´ ì£¼ë©´ ì§„í–‰ ì •ë„ë¥¼ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# í•™ìŠµì„ í•  ë•ŒëŠ” feature_tfidf ê°’ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
# cls.inertia_ ê°’ì„ inertia ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í•©ë‹ˆë‹¤.
for i in trange(start, end):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(feature_tfidf)
    inertia.append(kmeans.inertia_)
```

```python
# ìœ„ì—ì„œ êµ¬í•œ ê°’ì„ ì‹œê°í™”
# xì¶•ì—ëŠ” í´ëŸ¬ìŠ¤í„°ì˜ ìˆ˜ë¥¼ yì¶•ì—ëŠ” inertia ê°’ì„ ë„£ì–´ ê·¸ë¦¼

plt.plot(range(start, end), inertia)
plt.title("KMeans í´ëŸ¬ìŠ¤í„° ìˆ˜ ë¹„êµ")
```

```python
# n_clusters ì— ì ì ˆí•œ ê°’ì„ ë„£ì–´ì¤Œ
# fit.predict ë¥¼ í•˜ê³  ê²°ê³¼ë¥¼ cluster ë¼ëŠ” ìƒˆë¡œìš´ ì»¬ëŸ¼ì— ë‹´ìŒ
n_clusters = 60
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(feature_tfidf)
prediction = kmeans.predict(feature_tfidf)
df["cluster"] = prediction
```

```python
# df["cluster"] ì˜ ë¹ˆë„ìˆ˜ë¥¼ value_countsë¡œ ì„¸ì–´ë´…ë‹ˆë‹¤.
df["cluster"].value_counts().head(10)
```

```python
# batch_size ë¥¼ ì“¸ ìˆ˜ ìˆëŠ” MiniBatchKMeans ë¡œ êµ°ì§‘í™”
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
b_inertia = []
silhouettes = []

# ì ì ˆí•œ í´ëŸ¬ìŠ¤í„°ì˜ ê°¯ìˆ˜ë¥¼ ì•Œê¸° ìœ„í•´ inertia ê°’ì„ êµ¬í•¨
# trange ë¥¼ í†µí•´ ì‹œì‘ê³¼ ë ê°’ì„ ì§€ì •í•´ ì£¼ë©´ ì§„í–‰ ìƒíƒœë¥¼ í™•ì¸ í•  ìˆ˜ ìˆìŒ
# b_inertia ë¦¬ìŠ¤íŠ¸ì— cls.inertia_ ê°’ì„ ë„£ì–´ì¤ë‹ˆë‹¤.
for i in trange(start, end):
    mkmeans = MiniBatchKMeans(n_clusters=i, random_state=42)
    mkmeans.fit(feature_tfidf)
    b_inertia.append(mkmeans.inertia_)
    silhouettes.append(silhouette_score(feature_tfidf, mkmeans.labels_))
```

```python
# ìœ„ì—ì„œ êµ¬í•œ ê°’ì„ ì‹œê°í™” í•©ë‹ˆë‹¤.
# xì¶•ì—ëŠ” í´ëŸ¬ìŠ¤í„°ì˜ ìˆ˜ë¥¼ yì¶•ì—ëŠ” b_inertia ê°’ì„ ë„£ì–´ ì‹œê°í™”
plt.plot(range(start, end), b_inertia)
plt.title("MiniBatchKMeans í´ëŸ¬ìŠ¤í„° ìˆ˜ ë¹„êµ")
```

```python
plt.figure(figsize=(15, 4))
plt.title('Silhouette Score')
plt.plot(range(start, end), silhouettes)
plt.xticks(range(start, end))
plt.show()
```

```python
# yellowbrick.cluster ì—ì„œ KElbowVisualizer ë¶ˆëŸ¬ì˜¤ê¸°
from yellowbrick.cluster import KElbowVisualizer

KElbowM = KElbowVisualizer(kmeans, k=(start, end))
KElbowM.fit(feature_tfidf.toarray())
KElbowM.show()
```

```python
# n_clusters ìœ„ì—ì„œ ì •ì˜í•œ í´ëŸ¬ìŠ¤í„° ìˆ˜ë¥¼ ì‚¬ìš©
feature_array = feature_vector.toarray()
# ì˜ˆì¸¡í•œ í´ëŸ¬ìŠ¤í„°ì˜ ìœ ë‹ˆí¬ ê°’
labels = np.unique(prediction)
df_cluster_score = []
df_cluster = []
for label in labels:
    id_temp = np.where(prediction==label) # ì˜ˆì¸¡í•œ ê°’ì´ í´ëŸ¬ìŠ¤í„° ë²ˆí˜¸ì™€ ë§¤ì¹˜ ë˜ëŠ” ê²ƒì„ ê°€ì ¸ì˜´
    x_means = np.mean(feature_array[id_temp], axis = 0) # í´ëŸ¬ìŠ¤í„°ì˜ í‰ê·  ê°’ì„ êµ¬í•¨
    sorted_means = np.argsort(x_means)[::-1][:n_clusters] # ê°’ì„ ì—­ìˆœìœ¼ë¡œ ì •ë ¬í•´ì„œ í´ëŸ¬ìŠ¤í„° ìˆ˜ ë§Œí¼ ê°€ì ¸ì˜´
    features = vectorizer.get_feature_names_out()
    best_features = [(features[i], x_means[i]) for i in sorted_means] 
    # í´ëŸ¬ìŠ¤í„°ë³„ ì „ì²´ ìŠ¤ì½”ì–´
    df_score = pd.DataFrame(best_features, columns = ['features', 'score'])
    df_cluster_score.append(df_score)
    # í´ëŸ¬ìŠ¤í„° ëŒ€í‘œ í‚¤ì›Œë“œ
    df_cluster.append(best_features[0])
```

```python
# ê°œë³„ í´ëŸ¬ìŠ¤í„°ì—ì„œ ì ìˆ˜ê°€ ê°€ì¥ ë†’ì€ ë‹¨ì–´ë¥¼ ì¶”ì¶œ ì•„ë˜ ì ìˆ˜ê°€ í´ìˆ˜ë¡ ì˜ˆì¸¡ ì •í™•ë„ê°€ ë†’ìŒ
# MiniBatchKMeans ë¡œ ì˜ˆì¸¡í•œ ê°’ ê¸°ì¤€
pd.DataFrame(df_cluster, columns = ['features', 'score']).sort_values(by=["features", "score"], ascending=False)
```
